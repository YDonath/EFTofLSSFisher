```python
import os
from getdist import loadMCSamples, plots, mcsamples
import matplotlib.pyplot as plt
```

```python
def load_chain(chainname, basedir='./'): 
    chains = os.path.join(basedir, chainname, chainname)
    samples = loadMCSamples(chains, settings={'ignore_rows':0});
    return samples
```

```python
samples = load_chain('D4_b0', basedir='quad/06/P_B0')
```

```python
cosmo = ['Omega_m', 'h', 'sigma8'] 
g = plots.get_subplot_plotter(width_inch=5) 
g.triangle_plot([samples],
                cosmo, filled=True, 
                title_limit=1,
                legend_labels=[r'$P_\ell+B_0$ ']
               )
# g.export('paper_fig/tri_boss_b0.pdf')
plt.show()
```
