```python
def getfnlplot(base,save=0):
    locerror=getfnlerrors(base,["fNLloc"]+biasindexstfix,["fNLloc"])
    eqerror=getfnlerrors(base,["fNLeq"]+biasindexstfix,["fNLeq"])
    ortherror=getfnlerrors(base,["fNLorth"]+biasindexstfix,["fNLorth"])

    errors=np.array([locerror,eqerror,ortherror]).T.tolist()

    mean = [0,0,0]
    covTree=[[locerror[0]**2,0,0],[0,eqerror[0]**2,0],[0,0,ortherror[0]**2]]
    covLoop=[[locerror[1]**2,0,0],[0,eqerror[1]**2,0],[0,0,ortherror[1]**2]]
    covPrior=[[locerror[2]**2,0,0],[0,eqerror[2]**2,0],[0,0,ortherror[2]**2]]

    TreeG=gaussian_mixtures.GaussianND(mean, covTree,names=fnlnames)
    LoopG=gaussian_mixtures.GaussianND(mean, covLoop,names=fnlnames)
    PriorG=gaussian_mixtures.GaussianND(mean, covPrior,names=fnlnames)
    g = plots.get_subplot_plotter()
    g.plots_1d([TreeG,LoopG,PriorG], nx=3, share_y=True, legend_labels = fnllabs,colors=['tab:grey', 'tab:red','tab:blue'],legend_ncol =3,legend_loc ='right')
    if(save):
        plt.savefig(save+'.pdf',bbox_inches = "tight")
        with open(save+'.txt', 'w') as f:
            f.write(tabulate(gettable(around2(np.array(errors),2).tolist(),fnlnames,fnllabs),tablefmt="latex")) 
getfnlplot([MMpPBanamufullzT,MMpPBanamufullz,MMpPBanamufullz+MMpPBstrong],'plot/MMp/Priorsfnltest')
```
